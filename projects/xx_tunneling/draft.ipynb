```python
import altair
import numpy
import pandas
```

```python
def eckart_potential(nu_ts: float, e_forw: float, e_rev: float):
    a = e_forw - e_rev
    b = (numpy.sqrt(e_forw) + numpy.sqrt(e_rev)) ** 2

    def kappa(s: float) -> float:
        f = numpy.exp(s * numpy.abs(nu_ts) * numpy.sqrt(b / 2))
        return (f / (1 + f)) * (a + b / (1 + f))

    return kappa


s = numpy.linspace(-5, 5, 500)
kappa = eckart_potential(1/5, 40, 50)

source = pandas.DataFrame({"s": s, "kappa(s)": kappa(s)})

altair.Chart(source).mark_line().encode(x="s", y="kappa(s)")
```
